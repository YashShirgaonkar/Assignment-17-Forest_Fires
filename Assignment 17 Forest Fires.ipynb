# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
df1 = data.copy()
```

```python
df1 = df1.drop(['month','day'], axis=1)
```

```python
df1.head()
```

```python
x = df1.drop('size_category', axis=1)
y = df1['size_category']
```

```python
#Normalization
def norm(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
x = norm(x)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, stratify=y)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Training SVM model for Classification

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
```

### Model with linear kernel

```python
model_linear = SVC(kernel='linear')
```

```python
model_linear.fit(x_train,y_train)
```

```python
pred_train_linear = model_linear.predict(x_train)
```

```python
accuracy_score(y_train, pred_train_linear)
```

```python
confusion_matrix(y_train, pred_train_linear)
```

### Model with ploy kernel

```python
model_poly = SVC(kernel='poly')
```

```python
model_poly.fit(x_train,y_train)
```

```python
pred_train_poly = model_linear.predict(x_train)
```

```python
accuracy_score(y_train, pred_train_poly)
```

```python
confusion_matrix(y_train, pred_train_poly)
```

### Model with rbf kernel

```python
model_rbf = SVC(kernel='rbf')
```

```python
model_rbf.fit(x_train, y_train)
```

```python
pred_train_rbf = model_rbf.predict(x_train)
```

```python
accuracy_score(y_train, pred_train_rbf)
```

```python
confusion_matrix(y_train, pred_train_rbf)
```

### Model with sigmoid kernel

```python
model_sigmoid = SVC(kernel='sigmoid')
```

```python
model_sigmoid.fit(x_train, y_train)
```

```python
pred_train_sigmoid = model_sigmoid.predict(x_train)
```

```python
accuracy_score(y_train, pred_train_sigmoid)
```

```python
confusion_matrix(y_train, pred_train_sigmoid)
```

The Linear kernel has the most accuracy, so we will go ahead with that.
